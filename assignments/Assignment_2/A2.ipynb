```python
import pandas as pd
df = pd.read_csv("/content/admission.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt


df.hist(bins=30, color='r', align='right')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.suptitle("Histogram for Admission Dataset")
plt.tight_layout()
plt.show()
```
