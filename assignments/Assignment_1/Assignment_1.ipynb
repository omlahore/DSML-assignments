```python
#by using column name
df["col1"]
df[["col1","col2","col3"]]
```

```python
#By using row index

df.loc[["row3","row4","row5"]]
```

```python
#by using rowindex and columnindex[.iloc]

df.iloc[3:5,2:4]
```
