```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
insurance_df = pd.read_csv("/content/insurance.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=insurance_df, x='age', y='charges', hue='smoker', palette='Set1')
plt.title('Scatter Plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=insurance_df, x='region', y='charges')
plt.title('Total Charges by Region')
plt.xlabel('Region')
plt.ylabel('Total Charges')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=insurance_df, x='children', y='charges')
plt.title('Box Plot of Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Charges')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=insurance_df, x='bmi', bins=30)
plt.title('Histogram of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.regplot(x='age', y='charges', data=insurance_df)
plt.title('Regplot')
plt.show()
```

```python
sns.jointplot(x='age', y='charges', data=insurance_df, kind='scatter')
plt.title('Joint Plot of Age vs Charges')
plt.show()
```
